# Generate Random Latitudes & Longitudes

## Different kinds of random

```python
# To get a random floting number betwen 0 and 1.0
# random.random()
# To get a randmom integer, we could get the lower & upper boundery nb
# random.randint(-10,10)
# To get an integer that falls within a specifified range, but there's steps between bounderies 
# random.randrange(-9,9,2)
# To get a decimal that falls within a specifified range
# random.uniform(-10,10)
# We can use the round() to limit the number of decimals
# round(random.uniform (-10,10),2)
```

### To get 10 latitudes using the random() Function

```python
# To get 10 numbers 
x = 1
latitudes = []
while x < 11:
    random_lat = random.randint(-90, 89) + random.random()
    latitudes.append(random_lat)
    x += 1
    
latitudes    
```

### To get 10 Longitudes  using the random() Function

```python
x = 1
longitudes = []
while x < 11:
    random_lon = random.randint(-180, 179) + random.random()
    longitudes.append(random_lon)
    x += 1
    
longitudes 
```

### To get 10 Latitudes  using the randrange() Function

```python
x = 1
latitudes = []
while x < 11:
    random_lat = random.randrange(-90, 89, step=1) + random.random()
    latitudes.append(random_lat)
    x += 1
    
latitudes 
```

### To get 10 Latitudes using the uniform() Function



```python
x = 1
latitudes = []
while x < 11:
    random_lat = random.uniform(-90, 90)
    latitudes.append(random_lat)
    x += 1
    
latitudes 
```

### NumPy and random Modules
To generate more than a thousand latitudes and longitudes is to chain the NumPy module to the random module to create an array of latitudes or longitudes between the lowest and highest values, or -90° and 90°, and –180° and 180°

```python
# To generate a floating-point decimal number between -90.000 and 90.000
np.random.uniform(-90,90)
```

```python
# To generate 50 numbers 
np.random.uniform(-90.000, 90.000, size=50)
```
